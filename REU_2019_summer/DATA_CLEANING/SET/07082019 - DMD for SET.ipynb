## Creating Dataset for DMD

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
## Import data set

df = pd.read_csv('SET data archive 3416 cases recd thru end of 03 2019_cleaned_general.csv')
df.head()
```

```python
df['S10'].value_counts()
```

```python
col_kept = ['UniqueID']
col_kept.extend(df.loc[:,'intake':'retest'].columns.tolist())
col_kept.extend(df.loc[:,'GRYD_Zone_num':'INTERVIEW_DATE_FB'].columns.tolist())
col_kept.extend(df.loc[:,'ENROLLMENT_DATE_1':'SEX_1'].columns.tolist())
```

```python
# S section
que = df.loc[:,'S1':'S3'].columns.tolist()
que.extend(df.loc[:,'S4x':'S6ax'].columns.tolist())
que.extend(df.loc[:,'S9':'S10'].columns.tolist())
que.extend(df.loc[:,'S10x':'S11a'].columns.tolist())
que.extend(df.loc[:,'S11b':'S16'].columns.tolist())

# ME,P,C section
que.extend(df.loc[:,'ME17':'P26'].columns.tolist())
que.extend(['C1a','C2a','C3a','C4a','C5a'])

# FS section
que.extend(df.loc[:,'FS89_foster_now':'FS89_alone_past'].columns.tolist())

# F section
f_que.extend(df.loc[:,'F1':'F3'].columns.tolist())
f_que.extend(df.loc[:,'F4':'FS98'].columns.tolist())
f_que.extend(df.loc[:,'FS100':'F17x'].columns.tolist())
f_que.extend(df.loc[:,'F18':'F31b'].columns.tolist())
que.extend(f_que)

# G section
g_que = df.loc[:,'G2':'G3a'].columns.tolist()
g_que = df.loc[:,'EG1':'G9'].columns.tolist()
g_que.extend(df.loc[:,'G10x':'G23x'].columns.tolist())
g_que.extend(df.loc[:,'G24':'G37'].columns.tolist())
que.extend(g_que)

# X section
que.extend(df.loc[:,'X18':'X22'].columns.tolist())

# O section
que.extend(['A1','A2','A3'])
o_que = df.loc[:,'O4':'O9'].columns.tolist()
o_que.extend(df.loc[:,'O10':'O15'].columns.tolist())
o_que.extend(df.loc[:,'O16':'O24'].columns.tolist())
que.extend(o_que)

# X section
que.extend(df.loc[:,'X10':'X17'].columns.tolist())
```
