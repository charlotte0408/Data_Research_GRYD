## Y1:R1

### Male

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statistics 

#import 

Y1R1M_changeS = pd.read_csv('Y1_R1_M_eigenvector.csv')
Y1R1M_imptS = pd.read_csv('Y1_R1_M_importance.csv')
```

```python
Y1R1M_changeS = Y1R1M_changeS['0']
```

```python
Y1R1M_imptS = Y1R1M_imptS['0']
```

```python
Y1R1M_changeS = abs(Y1R1M_changeS)
Y1R1M_imptS = abs(Y1R1M_imptS)
```

```python
n = ['a1', 'a2',
       'a3', 'a4', 'a5', 'a6', 'b7', 'b8',
       'b9', 'c10', 'c11', 'c12', 'c13',
       'c14', 'c15', 'c16', 'de17', 'de18',
       'de19', 'de20', 'f21', 'f22', 'f23o',
       'f24', 'f25', 'f26', 'g27', 'g28',
       'g29', 'g30', 'g31', 'h32', 'h33',
       'h34', 'h35', 'h36', 'h37', 't38',
       't39']
```

```python
#Y1R1M_changeS

#take the mean
mean1 = statistics.mean(Y1R1M_changeS)
#find standard deviation
stdev1 = statistics.stdev(Y1R1M_changeS)
```

```python
#standardize data
for i in range(len(Y1R1M_changeS)):
    Y1R1M_changeS[i] = (Y1R1M_changeS[i] - mean1) / stdev1 
```

```python
#Y1R1M_imptS

#take the mean
mean2 = statistics.mean(Y1R1M_imptS)
#find standard deviation
stdev2 = statistics.stdev(Y1R1M_imptS)
```

```python
#standardize data
for i in range(len(Y1R1M_imptS)):
    Y1R1M_imptS[i] = (Y1R1M_imptS[i] - mean2) / stdev2 
```

```python
plt.figure(figsize=(20,15))
plt.scatter(Y1R1M_changeS, Y1R1M_imptS)

for i, txt in enumerate(n):
    plt.annotate(txt, (Y1R1M_changeS[i], Y1R1M_imptS[i]))

plt.axhline(0, color='red')
plt.axvline(0, color='red')
plt.title('Male Y1R1: Change vs Importance')
plt.xlabel('Standardized Weight of Entries in Dominant Eigenvector')
plt.ylabel('Standardized Weight of Entries in Last Row of A Matrix')
plt.show()
```
