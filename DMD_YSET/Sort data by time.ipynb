```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

df_c= pd.read_csv("yrr_sec.csv")
```

```python
df_c.head(10)
```

```python
df_c = df_c.drop(df_c.columns[0],axis=1)
```

```python
df_c.shape[0]
```

```python
def pos(df,num):
    if df.at[num,'UniqueID']==df.at[num+1,'UniqueID']\
        and df.at[num,'UniqueID']==df.at[num+2,'UniqueID']:
        return True
    else: 
        return False


intersec_y1=[]
intersec_r1=[]
intersec_r2=[]
intersect_y1_r1_r2=[]

for i in range(df_c.shape[0] - 2):
    if pos(df_c,i):
        intersec_y1.append(i)
        intersec_r1.append(i+1)
        intersec_r2.append(i+2)
        intersect_y1_r1_r2.append(i)
        intersect_y1_r1_r2.append(i+1)
        intersect_y1_r1_r2.append(i+2)
        
df_new=df_c.ix[intersect_y1_r1_r2]
df_new.to_csv('Intersection_Y1R1R2_0702Y1_R1_R2_on_0702clean_data.csv',index=False)
```
