```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Intersection_Y1R1R2_0702Y1_0702.csv')
dfrY1 = df.loc[:, 'a1_combo':'t39_combo']
Y1 = dfrY1.convert_objects(convert_numeric=True).values
Y1.shape
```

```python
List_index=[]
for i in range(2581):
    for j in range(39):
        If_true=True
        if np.isnan(Y1[i][j]):
            for k in range(len(List_index)):
                if i==List_index[k]:
                    If_true=False
            if If_true:
                List_index.append(i)
List_index           
```

```python
df = pd.read_csv('Intersection_Y1R1R2_0702R1_0702.csv')
dfrR1 = df.loc[:, 'a1_combo':'t39_combo']
R1 = dfrR1.convert_objects(convert_numeric=True).values
```

```python
for i in range(2581):
    for j in range(39):
        If_true=True
        if np.isnan(R1[i][j]):
            for k in range(len(List_index)):
                if i==List_index[k]:
                    If_true=False
            if If_true:
                List_index.append(i)
List_index           
```

```python
df = pd.read_csv('Intersection_Y1R1R2_0702R2_0702.csv')
dfrR2 = df.loc[:, 'a1_combo':'t39_combo']
R2 = dfrR2.convert_objects(convert_numeric=True).values
```

```python
for i in range(2581):
    for j in range(39):
        If_true=True
        if np.isnan(R2[i][j]):
            for k in range(len(List_index)):
                if i==List_index[k]:
                    If_true=False
            if If_true:
                List_index.append(i)
List_index.sort()
List_index      
```

```python
k=0
for m in range(len(List_index)):
    Y1=np.delete(Y1, (List_index[m]-k), 0)
    k=k+1
Y1=Y1.transpose()
Y1.shape
```

```python
k=0
for m in range(len(List_index)):
    R1=np.delete(R1, (List_index[m]-k), 0)
    k=k+1
R1=R1.transpose()
R1.shape
```

```python
k=0
for m in range(len(List_index)):
    R2=np.delete(R2, (List_index[m]-k), 0)
    k=k+1
R2.shape   
R2=R2.transpose()
R2.shape
```

```python
U_Y1,Sig_Y1,V_Y1=np.linalg.svd(Y1, full_matrices=False)    
U_Y1_T=U_Y1.conjugate().transpose()
V_Y1_T=V_Y1.conjugate().transpose()
Sig_inv_Y1=np.zeros((39, 39))
for i in range(39):
    for j in range(39):
        if i==j:
            Sig_inv_Y1[i][j]=1/Sig_Y1[i]
```

```python
U_R1,Sig_R1,V_R1=np.linalg.svd(R1, full_matrices=False)
U_R1_T=U_R1.conjugate().transpose()
V_R1_T=V_R1.conjugate().transpose()
Sig_inv_R1=np.zeros((39, 39))
for i in range(39):
    for j in range(39):
        if i==j:
            Sig_inv_R1[i][j]=1/Sig_R1[i]
```

```python
A_step1=np.dot(R1,V_Y1_T)
A_step2=np.dot(A_step1,Sig_inv_Y1)
A_Y1_R1=np.dot(A_step2,U_Y1_T)
A_Y1_R1
```

```python
A_step1=np.dot(R2,V_R1_T)
A_step2=np.dot(A_step1,Sig_inv_R1)
A_R1_R2=np.dot(A_step2,U_R1_T)
A_R1_R2
```
